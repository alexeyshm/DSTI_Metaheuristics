Function is convex and has only one global minimum. Therefore, the best option will be to use a gradiend based method

```python
def minimizeSphere(D):
    iters = pd.DataFrame(columns =['x', 'y'])
    
    def Callback(xi):
        step = len(iters)
        iters.loc[step, 'x'] = xi
        iters.loc[step, 'y'] = shiftedSphere(xi)    
        
    x0 = np.random.uniform(-10, 10, D)
    
    start = time.time()
    minimize(shiftedSphere, x0, method='BFGS', callback=Callback)
    
    print("Computational time: " + str(time.time() - start))
    
    print(iters)
    
    plt.plot(iters['y'])
```
